```python
# IDs repetidos
print(df['customerID'].duplicated().sum()) # Código único
```

```python
# Información demográfica
print(df['gender'].value_counts()) # Genero
print(df['SeniorCitizen'].value_counts()) # Edad (Mayor)
print(df['Partner'].value_counts()) # Pareja
print(df['Dependents'].value_counts()) # Dependientes
```

```python
# Información de servicios contratados
print(df['PhoneService'].value_counts()) # Teléfono
print(df['MultipleLines'].value_counts()) # Líneas múltiples
print(df['InternetService'].value_counts()) # Internet
print(df['OnlineSecurity'].value_counts()) # Seguridad
print(df['OnlineBackup'].value_counts()) # Copia de seguridad
print(df['DeviceProtection'].value_counts()) # Protección de dispositivos
print(df['TechSupport'].value_counts()) # Soporte técnico
print(df['StreamingTV'].value_counts()) # TV streaming
print(df['StreamingMovies'].value_counts()) # Películas streaming
```

```python
# Información de cuenta
print(df['tenure'].value_counts()) # Antiguedad
print(df['Contract'].value_counts()) # Contrato
print(df['PaperlessBilling'].value_counts()) # Facturación electrónica
print(df['PaymentMethod'].value_counts()) # Método de pago
print(df['MonthlyCharges'].value_counts()) # Cargos mensuales
print(df['TotalCharges'].value_counts()) # Cargos totales
```
